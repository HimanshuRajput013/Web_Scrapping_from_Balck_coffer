```python
import requests
from bs4 import BeautifulSoup
import os

# Define a function to extract data from a URL and save it to a text file
def get_text(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')
            # Extract text content from the webpage
            content=soup.findAll(attrs={'class':'td-post-content'})#extracting only text part
            text_content=content[0].text.replace('\xa0',"  ").replace('\n',"  ")#replace end line symbol with space
            # Extract text content from the webpage
            #text_content = soup.get_text()
            title=soup.findAll(attrs={'class':'entry-title'})#extracting title of website
            title=title[16].text.replace('\n',"  ").replace('/',"")
            text = title + text_content

            return text
        else:
            print(f"Failed to retrieve webpage {url} (URL ID: {url_id}) - Status code: {response.status_code}")
    except Exception as e:
        print(f"Error processing URL {url} (URL ID: {url_id}): {str(e)}")

# Itteration for all column
df['Text_Data'] = ''


# Iterate over each row in the DataFrame
for i in range(len(df)):
    # Extract text from the URL in the 'URL' column
    text = get_text(df['URL'][i])
    df.at[i, 'Text_Data'] = text

    url_id = df['URL_ID'][i]
    if text is not None:  # Check if text is not None
        filename = f"{url_id}.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)
            print(f"Text data saved to {filename}")
    else:
        print(f"Failed to retrieve text data for {url_id}")
```

```python
import string

# Define the function to clean text
def clean_text(text):
    # Check if text is not None
    if text is None:
        return ''  # or any other appropriate action

    # Define a set of stopwords and punctuation marks

    punctuation = set(string.punctuation)

    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation marks
    text = ''.join(char for char in text if char not in punctuation)

    # Tokenize the text
    tokens = text.split()

    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]

    # Join the tokens back into text
    cleaned_text = ' '.join(tokens)

    return cleaned_text

# Apply the clean_text function to each row in the 'Text_Data' column
df['Cleaned_Text'] = df['Text_Data'].apply(clean_text)
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize

# Function to calculate scores
def calculate_scores(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Calculate positive score
    positive_score = sum(1 for word in tokens if word in positive_words)

    # Calculate negative score
    negative_score = sum(1 for word in tokens if word in negative_words)

    # Calculate polarity score
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)

    # Calculate subjectivity score
    total_words = len(tokens)
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)

    return positive_score, negative_score, polarity_score, subjectivity_score

# Apply the function to each row in the DataFrame and store the results in new columns
df['Positive_Score'], df['Negative_Score'], df['Polarity_Score'], df['Subjectivity_Score'] = zip(*df['Cleaned_Text'].apply(calculate_scores))

# Display the DataFrame with the new columns
df
```

```python
import re

def calculate_num_sentences(text):
    # Check if the input is a string or bytes-like object
    if isinstance(text, str):
        # Remove punctuations
        text = re.sub(r'[^\w\s.]', '', text)

        # Split the text into sentences
        sentences = text.split('.')

        # Total number of sentences
        num_sentences = len(sentences)

        return num_sentences
    else:
        return 0  # or any other appropriate action

# Apply the function to each row in the DataFrame and store the results in a new column
df['Num_Sentences'] = df['Text_Data'].apply(calculate_num_sentences)

# Display the DataFrame with the new column
df
```

```python
def calculate_num_words(text):
    # Split the text into words
    words = text.split()

    # Total number of words
    num_words = len(words)

    return num_words

# Apply the function to each row in the DataFrame and store the results in a new column
df['Num_Words'] = df['Cleaned_Text'].apply(calculate_num_words)

# Display the DataFrame with the new column
df
```

```python
# Calculate average sentence length
df['Avg_Sentence_Length'] = df['Num_Words'] / df['Num_Sentences']

# Display the DataFrame with the new column
df
```

```python
nltk.download('cmudict')

# Load CMU Pronouncing Dictionary

d = cmudict.dict()

# Function to count syllables in a word
def syllable_count(word):
    if word.lower() in d:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]])
    else:
        return 0

# Function to count complex words in a text
def count_complex_words(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Count the number of complex words (words with more than two syllables)
    complex_word_count = sum(1 for word in words if syllable_count(word) > 2)

    return complex_word_count

df['Complex_Words'] = df['Cleaned_Text'].apply(count_complex_words)
df['Percentage_of_Complex_Words'] = df['Complex_Words'] / df['Num_Words']

# Calculate the Fog Index
df['Fog_Index'] = 0.4 * (df['Avg_Sentence_Length'] + df['Percentage_of_Complex_Words'])
df['Avg_number_of_words_per_sentece'] = df['Num_Words'] / df['Num_Sentences']
```

```python
def count_syllables_per_word(text):
    # Split the text into words
    words = text.split()
    syllable_count_per_word = {}
    syllable_words = []

    # Define vowels
    vowels = 'aeiou'

    # Iterate through each word
    for word in words:
        # Check if the word ends with 'es' or 'ed', and exclude it from syllable counting
        if word.lower().endswith('es') or word.lower().endswith('ed'):
            continue

        # Count the number of vowels in the word
        syllable_count_word = sum(1 for letter in word if letter.lower() in vowels)

        # If the syllable count is greater than 0, add it to the dictionary and list
        if syllable_count_word > 0:
            syllable_count_per_word[word] = syllable_count_word
            syllable_words.append(word)

    # Return the results
    return syllable_count_per_word, syllable_words

df['syllable_count_perword'] = df['Cleaned_Text'].apply(count_complex_words)
```

```python
def count_personal_pronouns(text):
    # Define the pattern to match personal pronouns
    pattern = r'\b(I|we|my|ours|us)\b'

    # Compile the regular expression pattern
    regex = re.compile(pattern, flags=re.IGNORECASE)

    # Find all matches of the pattern in the text
    matches = regex.findall(text)

    # Count the number of matches
    count = len(matches)

    return count

# Apply the function to each row in the DataFrame and store the results in a new column
df['Personal_Pronouns_Count'] = df['Cleaned_Text'].apply(count_personal_pronouns)
```

```python
def average_word_length(text):
    words = text.split()  # Split the text into words
    total_characters = sum(len(word) for word in words)  # Sum of the lengths of all words
    num_words = len(words)  # Total number of words
    return total_characters / num_words if num_words != 0 else 0  # Calculate average word length

# Apply the average_word_length function directly to the 'Text_Data' column
df['Average_Word_Length'] = df['Cleaned_Text'].apply(average_word_length)
```
